# Unsupervised Analysis (K-Means)
Este notebook usa un dataset sintético generado automáticamente.

```
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data
X = pd.read_csv('synthetic_training_data.csv')

features=['income','age','loan_amount','tenure_months','late_payments']
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X[features])

km=KMeans(n_clusters=4,random_state=0).fit(X_scaled)
X['cluster']=km.labels_
X.head()
```

```
X.groupby('cluster')[features].mean()
```
